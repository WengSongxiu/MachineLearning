## 读取数据

```python
data = pd.read_csv('../data/insurance_claims.csv')
target_name = 'fraud_reported'
target = data[target_name]
feature = data.drop([target_name], axis=1)

# 数据的一些特殊处理
target.replace(to_replace='Y', value=1, inplace=True)
target.replace(to_replace='N', value=0, inplace=True)
feature.drop(['_c39'],inplace=True,axis=1)
numerical_feature = feature.select_dtypes(exclude=['object']).columns
categorical_feature = feature.select_dtypes(include=['object']).columns
```

## 5.分析特征跟标签的相关性（热力图+皮尔森相关系数）

```python
def data_correlation(feature):
    k = 10  # number of variables for heatmap
    train_corr = feature.corr()
    cols = train_corr.nlargest(k, 'target')['target'].index
    cm = np.corrcoef(feature[cols].values.T)
    hm = plt.subplots(figsize=(10, 10))  # 调整画布大小
    hm = sns.heatmap(feature[cols].corr(), annot=True, square=True)
    plt.show()
```

```python
# 针对连续特征
data = feature
data['target'] = target
data_correlation(data)
```

```python
train_corr = feature.corr()
related_features = train_corr[abs(train_corr) > 0.5]
type(related_features)
related_features
```

```python
def pearson_corr(df, label):
    feature_corr = {}
    for f in df:
        df_tmp = df[df[f].notnull()]
        col = df_tmp[f].corr(df_tmp[label], method='pearson')
        feature_corr[f] = col
        
    corr_list = sorted(feature_corr.items(), key=lambda kv: (-kv[1], kv[0]))
    for t in corr_list:
        print(t)
```

```python
all_columns = list(numerical_feature)
all_columns.append('target')
pearson_corr(data[all_columns], 'target')
```

## 6.分析特征的分布情况（直方图-QQ图）

```python

train_cols = 6
train_rows = len(numerical_feature)
plt.figure(figsize=(4*train_cols,5*train_rows))

i=0

for col in numerical_feature:
    i+=1
    ax=plt.subplot(train_rows,train_cols,i)
    sns.distplot(data[col].dropna(),fit=stats.norm)
    i+=1
    ax=plt.subplot(train_rows,train_cols,i)
    res = stats.probplot(data[col], plot=plt)
plt.show()
```

```python
focus_categorical_feature = ['incident_state','policy_state','insured_education_level','insured_sex','insured_relationship','incident_type','insured_occupation','insured_hobbies','authorities_contacted','incident_severity','incident_type','policy_csl']
train_cols = 4
train_rows = len(focus_categorical_feature)
plt.figure(figsize=(4*train_cols,5*train_rows))
i=0
for col in focus_categorical_feature:
    i+=1
    ax = plt.subplot(train_rows,train_cols,i)
    ax = data.groupby(col).target.count().plot.bar(ylim=0)
    ax.set_ylabel('Fraud reported')
    
plt.show()
```

```python
focus_categorical_feature = ['incident_state','policy_state','insured_education_level','insured_sex','insured_relationship','incident_type','insured_occupation','insured_hobbies','authorities_contacted','incident_severity','incident_type','policy_csl']
plt.figure(figsize=(10,6))
for col in focus_categorical_feature:
    table=pd.crosstab(data[col], data.target)
    table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
    plt.title(col+' vs Fraud', fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Fraud Rate')
plt.show()
```

## 7.分析特征跟标签的回归关系（直方图-散点图）

```python
fcols = 4
frows = len(numerical_feature)
plt.figure(figsize=(5*fcols,4*frows))

i=0
for col in numerical_feature:
    i+=1
    ax=plt.subplot(frows,fcols,i)
    sns.regplot(x=col, y='target', data=data, ax=ax, 
                scatter_kws={'marker':'.','s':3,'alpha':0.3},
                line_kws={'color':'k'});
    plt.xlabel(col)
    plt.ylabel('target')
    
    i+=1
    ax=plt.subplot(frows,fcols,i)
    sns.distplot(data[col].dropna())
    plt.xlabel(col)
```
